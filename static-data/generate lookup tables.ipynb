# Generate lookup tables

The main script needs some lookup tables. They're generated here.

```python
import pandas as pd
```

## List of state abbreviations and populations

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_states_and_territories_of_the_United_States')
```

```python
states = df[0].iloc[:,[0,1,5]].copy()
states.columns = ['state_full', 'state_code', 'population']
states.loc[:,'state_full'] = states.state_full.str.split('[', n=1).str.get(0)
states
```

```python
dc = df[1].iloc[:,[0,1,3]].copy()
dc.columns = ['state_full', 'state_code', 'population']
dc.loc[:,'state_code'] = 'D.C.'
dc
```

```python
states.append(dc).to_csv('states.csv', index=False, encoding='utf-8')
```

## List of countries and population

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population')[0]
```

```python
countries = df.iloc[:,[1,2]].copy()
countries.columns = ['country', 'population']
countries.loc[:,'country'] = countries.country.str.split('[', n=1).str.get(0)
countries
```

```python
countries.to_csv('countries.csv', index=False, encoding='utf-8')
```

## List of Chinese provinces

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_Chinese_administrative_divisions_by_population')[3]
```

```python
china = df.iloc[:,[0,1]].copy()
china.columns = ['province', 'population']
china.loc[:,'province'] = china.province.str.split('[', n=1).str.get(0)
china
```

```python
china.to_csv('china.csv', index=False, encoding='utf-8')
```
